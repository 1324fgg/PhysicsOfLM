```python
#抽取P和S

def pAnds(result, sequences, parents):
    #p是对parent的纵向抽取，抽取result中每个元素的父亲，及父亲的父亲的位置，每个元素有6个祖宗，
    #如最后一个元素3，其祖宗的位置为[102, 34, 13, 5, 2, 1]，102表示sequence[5]的第102个元素为元素3的父亲，1表示第一层元素的父亲为根结点22
    p = []
    
    #s是根据p对sequence的抽取，保存result每个元素的父亲元素，例如最后一个元素3，其自己和祖宗为 [3, 7, 10, 15, 17, 20]，7为sequence[5]的第102个元素
    s = []
    
    # 生成父节点数组
    def generate_parents_array(result, sequences, parents):
        p = [[] for _ in range(len(result))]
        for i in range(len(result)):
            current_level = 6
            current_pos = i + 1
            while current_level > 0:
                if current_level in parents and current_pos <= len(parents[current_level]):
                    p[i].append(parents[current_level][current_pos - 1])
                    current_pos = parents[current_level][current_pos - 1]
                else:
                    p[i].append(None)
                current_level -= 1
        return p
    
    p = generate_parents_array(result, sequences, parents)
    
    # 根据p访问sequences生成s
    def generate_s_array(result, sequences, p):
        s = []
        for i in range(len(result)):
            current_level = 6
            current_pos = i + 1
            s_element = []
            while current_level > 0:
                if current_level in sequences and current_pos <= len(sequences[current_level]):
                    s_element.append(sequences[current_level][current_pos - 1])
                    current_pos = p[i][6 - current_level]
                current_level -= 1
            s.append(s_element)
        return s
    
    s = generate_s_array(result, sequences, p)
    
    
    print("Result:", result)
    print("Sequences:", sequences)
    print("Parents:", parents)
    print("P:", p)
    print("S:", s)
```

```python
import random

# 初始化保存信息的字典
#sequence保存生成路径，规则1生成sequence[1]，直到规则6生成sequence[6]
sequences = {}

#parent保存各层结果的父亲位置，如parent[2]保存sequence[2]：[16, 18, 17, 16, 17]的父亲位置为[1, 1, 1, 2, 2]
parents = {}

# 递归生成规则
def generate_sequence(n, level, rules, parent_pos):
    if level > 6 or n not in rules[level]:
        return [n]
    next_values = random.choice(rules[level][n])
    result = []
    if level not in sequences:
        sequences[level] = []
        parents[level] = []
    for i, value in enumerate(next_values):
        sequences[level].append(value)
        parents[level].append(parent_pos)
        result.extend(generate_sequence(value, level + 1, rules, len(sequences[level])))
    return result

# 批量生成结果
def batch_generate_results(initial_value, num_batches, rules):
    all_results = []
    for _ in range(num_batches):
        sequences.clear()
        parents.clear()
        result = generate_sequence(initial_value, 1, rules, 1)
        all_results.append({
            'result': result,
            'sequences': sequences.copy(),
            'parents': parents.copy()
        })
    return all_results

# 生成并存储批量结果
initial_value = 22
num_batches = 5  # 生成5个批次的结果
all_results = batch_generate_results(initial_value, num_batches, rules3b)

# 打印批量结果
for i, res in enumerate(all_results):
    print(f"Batch {i + 1}:")
    print("Result:", res['result'])
    print("Sequences:", res['sequences'])
    print("Parents:", res['parents'])
    print()
```
